```python
#If you want to check whether feature is guassian or normally distributed
#Q-Q plot
def plot_data(fraud,feature):
    plt.figure(figsize=(10,6))
    plt.subplot(1,1,2)
    fraud[feature].hist()
    plt.subplot(1,2,2)
    stat.probplot(fraud[feature],dist='norm',plot=pylab)
    plt.show()
   
```

```python
plot_data(fraud,'Age')
```

```python
# Density plot for each attribute
fig, ax = plt.subplots(5,3, figsize=(14,12))
axes_ = [axes_row for axes in ax for axes_row in axes]
for i,c in enumerate(fraud.iloc[:,1:]):
    sns.distplot(fraud[c], ax = axes_[i], color = 'red')
    plt.tight_layout()
```

```python
# Box plot for each attribute
fig, ax = plt.subplots(5,3, figsize=(14,12))
axes_ = [axes_row for axes in ax for axes_row in axes]
for i,c in enumerate(fraud.columns):
    sns.boxplot(fraud[c], ax = axes_[i], color = 'green')
    plt.tight_layout()
```
